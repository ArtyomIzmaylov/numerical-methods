## Взаимосвязь переменных

### Взаимосвязь категориальной (пол) и количественной (рост) переменных

```python
# импортируем модуль pyplot из библиотеки matplotlib
import matplotlib.pyplot as plt
```

```python
# предположим мы спрашивали людей (на этот раз мужчин и женщин) об их росте

# на код в этой ячейке опять же не обращайте внимания, он имитирует случайные данные
import numpy as np
np.random.seed(42)
height_men = list(np.round(np.random.normal(180, 10, 1000)))
height_women = list(np.round(np.random.normal(160, 10, 1000)))
```

```python
# теперь у нас два набора данных: мужчины и женщины
print(height_men)
print(height_women)
```

```python
# выберем количество интервалов
bins = 15

# новая для нас возможность - размер графика
plt.figure(figsize = (10,10))

# построим две гистограммы
# alpha задает их прозрачность
plt.hist(height_men, bins, alpha = 0.5, label = 'Рост мужчин')
plt.hist(height_women, bins, alpha = 0.5, label = 'Рост женщин')

# укажем, где мы хотим расположить легенду графика и пропишем размер шрифта
plt.legend(loc = 'upper left', prop = {'size': 14})

# добавим подписи и укажем размер шрифта
plt.xlabel('Рост, см', fontsize = 16)
plt.ylabel('Количество респондентов', fontsize = 16)
plt.title('Распределение роста мужчин и женщин в России', fontsize = 18)
```

```python
# соберем данные для роста и обхвата шеи
height_women_new = [1.48, 1.49, 1.49, 1.50, 1.51, 1.52, 1.52, 1.53, 1.53, 1.54, 1.55, 1.56, 1.57, 1.57, 1.58, 1.58, 1.59, 1.60, 1.61, 1.62, 1.63, 1.64, 1.65, 1.65, 1.66, 1.67, 1.67, 1.68, 1.68,  1.69, 1.70, 1.70, 1.71, 1.71, 1.71, 1.74, 1.75, 1.76, 1.77, 1.77, 1.78]
neck_women = [29.1, 30.0, 30.1, 30.2, 30.4, 30.6, 30.8, 30.9, 31.0, 30.6, 30.7, 30.9, 31.0, 31.2, 31.3, 32.0, 31.4, 31.9, 32.4, 32.8, 32.8, 33.3, 33.6, 33.0, 33.9, 33.8, 35.0, 34.5, 34.7, 34.6, 34.2, 34.8, 35.5, 36.0, 36.2, 36.3, 36.6, 36.8, 36.8, 37.0, 38.5]
```

### Корреляция двух переменных

```python
# импортируем модуль stats из библиотеки scipy
import scipy.stats as st
# передаем два списка в качестве аргумента и выводим первый элемент результата с индексом [0]
# затем округляем результат
np.round(st.pearsonr(height_women_new, list(reversed(neck_women)))[0], 2)
```

## Построение модели

### Модель регрессии

```python
# переведем наши списки в особый формат, массив Numpy
X = np.array(height_women_new)
y = np.array(neck_women)

# зададим размер графика
plt.figure(figsize = (10, 6))

# снова построим точечную диаграмму
plt.scatter(X, y)

# создадим модель с помощью функции polyfit()
# slope и intercept - это наклон и сдвиг вверх-вниз нашей прямой
# и одновременно те самые веса модели w и e
slope, intercept = np.polyfit(X, y, 1)

# с их помощью построим график прямой
# каждой точке X будет соответствовать прогнозное значение X*slope + intercept
# r означает red, красный цвет линии
plt.plot(X, X * slope + intercept, 'r')

# добавим подписи
plt.xlabel('Рост женщин в России, м', fontsize = 16)
plt.ylabel('Обхват шеи, см', fontsize = 16)
plt.title('Зависимость роста и окружности шеи у женщин в России', fontsize = 18)
```

```python
print(np.round(slope, 2), np.round(intercept, 2))
```
